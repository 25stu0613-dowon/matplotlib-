```python
import matplotlib.pyplot as plt

# ÌïôÏÉù Ïàò
num_students = 5

# ÌïôÏÉù Ï†ïÎ≥¥ Ï†ÄÏû• Î¶¨Ïä§Ìä∏
students = []

# ÏûÖÎ†• Î∞õÍ∏∞
for i in range(num_students):
    print(f"\n{i+1}Î≤àÏß∏ ÌïôÏÉù Ï†ïÎ≥¥ ÏûÖÎ†•")
    name = input("Ïù¥Î¶Ñ: ")

    def get_score(subject):
        while True:
            try:
                score_str = input(f"{subject} Ï†êÏàò: ")
                score = int(score_str)
                return score
            except ValueError:
                print("ÏûòÎ™ªÎêú ÏûÖÎ†•ÏûÖÎãàÎã§. Ïà´ÏûêÎ•º ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî.")

    korean = get_score("Íµ≠Ïñ¥")
    math = get_score("ÏàòÌïô")
    science = get_score("Í≥ºÌïô")

    student = {
        "Ïù¥Î¶Ñ": name,
        "Íµ≠Ïñ¥": korean,
        "ÏàòÌïô": math,
        "Í≥ºÌïô": science,
        "Ìï©Í≥Ñ": korean + math + science,
        "ÌèâÍ∑†": round((korean + math + science) / 3, 2)
    }
    students.append(student)

# Í∞úÏù∏Î≥Ñ ÏÑ±Ï†Å Ï∂úÎ†•
print("\nüßë‚Äçüéì ÌïôÏÉùÎ≥Ñ ÏÑ±Ï†Å")
print("-" * 40)
for s in students:
    print(f"Ïù¥Î¶Ñ: {s['Ïù¥Î¶Ñ']}, Íµ≠Ïñ¥: {s['Íµ≠Ïñ¥']}, ÏàòÌïô: {s['ÏàòÌïô']}, Í≥ºÌïô: {s['Í≥ºÌïô']}, Ìï©Í≥Ñ: {s['Ìï©Í≥Ñ']}, ÌèâÍ∑†: {s['ÌèâÍ∑†']}")

# Í≥ºÎ™©Î≥Ñ Ìï©Í≥Ñ Î∞è ÌèâÍ∑† Í≥ÑÏÇ∞
total_korean = sum(s['Íµ≠Ïñ¥'] for s in students)
total_math = sum(s['ÏàòÌïô'] for s in students)
total_science = sum(s['Í≥ºÌïô'] for s in students)

avg_korean = round(total_korean / num_students, 2)
avg_math = round(total_math / num_students, 2)
avg_science = round(total_science / num_students, 2)

# Í≥ºÎ™©Î≥Ñ Ï¥ùÏ†ê Î∞è ÌèâÍ∑† Ï∂úÎ†•
print("\nüìä Í≥ºÎ™©Î≥Ñ Ï¥ùÏ†ê Î∞è ÌèâÍ∑†")
print("-" * 40)
print(f"Íµ≠Ïñ¥ - Ï¥ùÏ†ê: {total_korean}, ÌèâÍ∑†: {avg_korean}")
print(f"ÏàòÌïô - Ï¥ùÏ†ê: {total_math}, ÌèâÍ∑†: {avg_math}")
print(f"Í≥ºÌïô - Ï¥ùÏ†ê: {total_science}, ÌèâÍ∑†: {avg_science}")

# ‚úÖ ÏãúÍ∞ÅÌôî (matplotlib)
names = [s['Ïù¥Î¶Ñ'] for s in students]
scores_k = [s['Íµ≠Ïñ¥'] for s in students]
scores_m = [s['ÏàòÌïô'] for s in students]
scores_s = [s['Í≥ºÌïô'] for s in students]

plt.figure(figsize=(10,5))
plt.title("ÌïôÏÉùÎ≥Ñ Í≥ºÎ™© Ï†êÏàò ÎπÑÍµê")
plt.xlabel("ÌïôÏÉù")
plt.ylabel("Ï†êÏàò")
plt.plot(names, scores_k, marker="o", label="Íµ≠Ïñ¥")
plt.plot(names, scores_m, marker="o", label="ÏàòÌïô")
plt.plot(names, scores_s, marker="o", label="Í≥ºÌïô")
plt.legend()
plt.grid(True)
plt.show()
```
