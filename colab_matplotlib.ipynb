```python
import numpy as np
import matplotlib.pyplot as plt

# t 값 설정 (0부터 2π까지)
t = np.linspace(0, 2 * np.pi, 1000)

# 하트 모양을 만드는 x, y 좌표
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# 하트 그리기
plt.figure(figsize=(6, 6))
plt.plot(x, y, color='red')
plt.fill(x, y, color='red', alpha=0.6)  # 색상 채우기
plt.title("❤️ Heart with Matplotlib", fontsize=16)
plt.axis('off')  # 축 숨기기
plt.show()
```
